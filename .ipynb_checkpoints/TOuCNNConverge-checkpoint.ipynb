```python
from numpy import nan
import matplotlib.pyplot as plt
from CNNOptimizer import CNNOptimizer
```

# Utility functions

```python
def iter_print(x, history, force=False):
    iteration = len(history["Objective"])
    if iteration % 10 == 0 or force:
        title = (f'Iteration {iteration:03}:   '
                 f'J={history["Objective"][-1]:0.3f}     '
                 f'V={history["Volume"][-1]:0.3f}')
        plt.imshow(x, origin='lower', cmap='Greys', vmin=0, vmax=1, interpolation='nearest')
        plt.title(title)
        plt.tick_params(axis='both', which='both', colors='w')
        plt.show()
```

```python
def optimize(opt):
    print(f'Optimization started.')
    x, time = opt.optimize(iter_print)
    iter_print(x, opt.history, True)
    print(f'Optimization finished in {time:0.2f} seconds.')
    ax = plt.subplots(figsize=(10, 5))[1]
    ax.plot(opt.history['Objective'] / max(opt.history['Objective']), label='Scaled Objective')
    ax.plot(opt.history['Volume'],'--', label='Volume Fraction')
    ax.plot(opt.history['Gray'],':', label='Gray Elements')
    ax.set_yscale("symlog", linthresh=0.01)
    ax.legend()
    ax.set_title('Convergence history')
    plt.ylim(0, 1)
    plt.show()
```

# Default configuration

```python
options = {
    'use_gpu': True,
    'mesh_size': (256, 128),
    'alpha_increase': 1,
    'penalty': 2, 'penalty_increase': 0.025,
    'learning_rate': 2.5E-4,
    'min_it': 25, 'max_it': 500, 'converge_criteria': 0.02
}
```

# Tip-loaded cantilever example

```python
bc = [{'S': (0, 0), 'E': (0, 1), 'D': (0, 0), 'F': 0},
      {'S': (1, 0), 'E': (1, 0), 'D': (nan, nan), 'F': lambda x, y: (0, -1)}]
options['volume_fraction'] = 0.75
options['symmetry_axis'] = None
optimizer = CNNOptimizer(bc, [], options)
optimize(optimizer)
```
