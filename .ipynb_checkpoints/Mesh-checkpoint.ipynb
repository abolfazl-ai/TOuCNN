```python
import numpy as np
from numpy import nan
import matplotlib.pyplot as plt
from IPython.display import clear_output
from CNNOptimizer import CNNOptimizer
```

# Utility functions

```python
def iter_print(x, history):
    clear_output(wait=True)
    title = (f'Iteration {len(history["Objective"]):03}:   '
             f'J={history["Objective"][-1]:0.3f}     '
             f'V={history["Volume"][-1]:0.3f}     '
             f'G={history["Gray"][-1]:0.3f}')
    plt.imshow(x, origin='lower', cmap='Greys', vmin=0, vmax=1, interpolation='nearest')
    plt.title(title)
    plt.tick_params(axis='both', which='both', colors='w')
    plt.show()
```

```python
def optimize(opt):
    print(f'Optimization started.')
    x, time = opt.optimize(iter_print)
    print(f'Optimization finished in {time:0.2f} seconds.')
    return time / len(opt.history["Objective"])
```

# Default configuration

```python
elem_num, avg_time = [], []

bc = [{'S': (0, 0), 'E': (0, 1), 'D': (0, 0), 'F': 0},
      {'S': (1, 0.5), 'E': (1, 0.5), 'D': (nan, nan), 'F': lambda x, y: (0, -1)}]
options = {
    'use_gpu': True,
    'alpha_increase': 1,
    'volume_fraction': 0.5,
    'penalty': 2, 'penalty_increase': 0.05,
    'learning_rate': 1.0E-4,
    'symmetry_axis': None,
    'min_it': 25, 'max_it': 500, 'converge_criteria': 0.02
}
```

# Mesh size: 64x32

```python
options['mesh_size'] = (64, 32)
optimizer = CNNOptimizer(bc, [], options)
t = optimize(optimizer)
elem_num.append(np.prod(options['mesh_size']))
avg_time.append(t)
```
