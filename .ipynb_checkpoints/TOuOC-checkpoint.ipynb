```python
import matplotlib.pyplot as plt
from IPython.display import clear_output
from OCOptimizer import OCOptimizer
from examples import get_example
```

# Utility functions

```python
def iter_print(x, history):
    clear_output(wait=True)
    title = (f'Iteration {len(history["Objective"]):03}:   '
             f'J={history["Objective"][-1]:0.3f}     '
             f'V={history["Volume"][-1]:0.3f}')
    plt.imshow(x, origin='lower', cmap='Greys', vmin=0, vmax=1, interpolation='nearest')
    plt.title(title)
    plt.tick_params(axis='both', which='both', colors='w')
    plt.show()
```

```python
def optimize(opt):
    print(f'Optimization started.')
    x, time = opt.optimize(iter_print)
    print(f'Optimization finished in {time:0.2f} seconds.')
    ax = plt.subplots()[1]
    ax.plot(opt.history['Objective'] / max(opt.history['Objective']), label='Scaled Objective')
    ax.plot(opt.history['Volume'], label='Volume Fraction')
    ax.plot(opt.history['Change'], label='Change')
    ax.legend()
    plt.ylim(0, 1)
    plt.show() 
```

# Tip-loaded cantilever example

```python
_, bc, frozen, cnn_opts = get_example(1)
oc_opts = {
    'mesh_size': cnn_opts['mesh_size'],
    'volume_fraction': cnn_opts['volume_fraction'],
    'symmetry_axis': cnn_opts['symmetry_axis'],
    'penalty': 3.0, 'penalty_increase': 0.025,
    'move': 0.2, 'filter_radius': 3,
    'min_it': 25, 'max_it': 500, 'converge_criteria': 0.001
}
optimizer = OCOptimizer(bc, frozen, oc_opts)
optimize(optimizer)
```
